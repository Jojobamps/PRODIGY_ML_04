### IMPORTING LIBRARIES

```python
import os
import cv2
import numpy as np
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

### LOADING THE DATA 

```python
gesture_folders = []

for i in range(10):
    folder = r"C:\Users\josep\Desktop\PRODIGY_INFOTECH_TASK_4\leapGestRecog\0{}".format(i)
    gesture_folders.append(folder)

files_names = ['01_palm', '02_l', '03_fist', '04_fist_moved', '05_thumb', '06_index', '07_ok', '08_palm_moved', '09_c', '10_down']
gesture_folders
```

```python
training_data=[]

for folder in gesture_folders:
        Class_num=folder[-1]
    
        for file in files_names:
            path = os.path.join(folder, file)
            for img in tqdm(os.listdir(path)):
                img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
                
                training_data.append([img_array.flatten(),int(Class_num)])
```

```python
image_data=[]
labels=[]

for feature,label in training_data:
    image_data.append(feature)
    labels.append(label)
```
